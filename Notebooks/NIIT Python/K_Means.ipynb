## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Create Data

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(
    n_samples = 200,  #no. of rows
    n_features = 2,  #no. of columns
    centers = 4,  #No. of clusters in the original data
    cluster_std = 1.8,
    random_state = 99
)
```

```python
len(data)
# data[0] -> is the data we need
# data[1] is the original cluster label
# We will use data[1] for plotting but it will not be needed
# Because KMeans is an unsupervised algorithm, so labels are not needed
```

```python
data[0][:5]  #x1 and x2
```

```python
data[0].shape
```

```python
data[1][:5]
```

## Visualizing Data

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1], cmap='rainbow', edgecolor='k')
```

## KMeans Model

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4)
```

```python
model.fit(data[0])  #We don't need data[1]
```

```python
model.cluster_centers_  #Coordinates of cluster center
```

```python
model.labels_  #Cluster numbering as decided by KMeans algorithms
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10, 6))

axes[0].set_title('K Means')
axes[0].scatter(
data[0][:, 0], data[0][:, 1],
c = model.labels_, cmap = 'rainbow', edgecolor='k'
)
axes[0].scatter(   #Drawing the cluster centers
    model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],
    c='r', edgecolor='k', s=200, alpha=0.3
)

axes[1].set_title('Original')
axes[1].scatter(
    data[0][:, 0], data[0][:, 1],
    c = data[1], cmap='rainbow', edgecolor='k'
)
```

## Elbow Point

```python
inertia = []

for k in range(1, 11):
    model = KMeans(n_clusters=k)
    model.fit(data[0])
    inertia.append(model.inertia_)
    
# Inertia is a measure of the proximity of the data points on the clusters
# We want to minimize inertia but upto a certain point
```

```python
inertia
```

```python
plt.plot(range(1, 11), inertia, marker='o')
```

If the plot is like an arm, ideal value of k is the elbow

```python
model = KMeans(n_clusters=3)
model.fit(data[0])

fig, axes = plt.subplots(nrows=1, ncols=2, sharey=True, figsize=(10, 6))

axes[0].set_title('K Means')
axes[0].scatter(
data[0][:, 0], data[0][:, 1],
c = model.labels_, cmap = 'rainbow', edgecolor='k'
)
axes[0].scatter(   #Drawing the cluster centers
    model.cluster_centers_[:, 0], model.cluster_centers_[:, 1],
    c='r', edgecolor='k', s=200, alpha=0.3
)

axes[1].set_title('Original')
axes[1].scatter(
    data[0][:, 0], data[0][:, 1],
    c = data[1], cmap='rainbow', edgecolor='k'
)
```
