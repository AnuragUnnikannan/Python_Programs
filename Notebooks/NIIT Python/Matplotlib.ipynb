## Make Imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Generating Data

```python
x = np.arange(0, 10)
y = 2 * x
a = np.linspace(0, 10, 11)
b = a ** 4
print("x: ",x)
print("y: ",y)
print("a: ",a)
print("b: ",b)
```

## Simple Plot

```python
plt.plot(x, y)

#Title and Axis Labels
plt.title('Simple Graph')
plt.xlabel('X-Axis - x')
plt.ylabel('Y-Axis - y')

#optional for notebook
plt.show()
```

```python
plt.plot(x, y)
plt.title('Simple Graph')
plt.xlabel('X-Axis - x')
plt.ylabel('Y-Axis - y')

#limit the axes
plt.xlim(0, 7)
plt.ylim(4, 12)

#save th pot as image
plt.savefig('fig1.jpeg')
plt.show()
```

## Figure Object

```python
#create the figure
fig = plt.figure()

#add axes
axes = fig.add_axes([0, 0, 1, 1])
#1. Distance of the bottom left corner of the plot from left edge of figure (x-coordinate)
#2. Distance of the bottom left corner of the plot from bottom edge of figure (y-coordinate)
#3. Width of plot
#4. Height of plot

axes.plot(x, y)
plt.show()
```

```python
#we can specify the size of figure
# figsize = (width, height)  (both values will be in inches)
# dpi is dots per inch (pixels per inch)
# using both parameters, we can set the size in pixels

fig = plt.figure(figsize=(10, 5), dpi=200)

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, y)
axes1.set_title('Outer')

axes2 = fig.add_axes([0.2, 0.6, 0.4, 0.3])
axes2.plot(a, b)
axes2.set_title('Inner')

# to save it, use ig.savefig()
# size of the figure is actually the outermost rectangle
# without bbox_inches='tight', title, labels and x, y numbering will not be saved in picture
fig.savefig('fig2.png', bbox_inches='tight')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(x, y)
axes1.set_title('Outer')
axes1.set_xlabel('X-Axis')
axes1.set_ylabel('Y-Axis')

axes2 = fig.add_axes([1.1, 0, 0.4, 0.3])
axes2.plot(a,b)
axes2.set_title('Inner')

plt.show()
```

## Subplots

```python
#create 2 subplots side by side
fig, axes = plt.subplots(nrows=1, ncols=2)

#plotting both axes
axes[0].plot(a, b)
axes[1].plot(x, y)

#title and labels
axes[0].set_title('1st graph')
axes[0].set_xlabel('a')
axes[0].set_ylabel('b')
axes[1].set_title('2nd graph')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')

#setting title for the whole figure
fig.suptitle('Figure Title', fontsize=20)

#prevent overlap
fig.tight_layout()

plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 5), dpi=200)

#axes is a 2D numpy array
axes[0, 0].plot(a, b)
axes[0, 1].plot(x, y)
axes[1, 0].plot(x, x**3)
axes[1, 1].plot(x, np.sqrt(x))

#tight_layout() is used for adjusting spaces between graphs
fig.tight_layout()
plt.show()
```

## Multiple plots in single axis

```python
fig, axes = plt.subplots()

axes.plot(x, y, label='x vs y')
axes.plot(x, y**2, label='x vs y^2')
axes.plot(x, y**3, label='x vs y^3')

axes.legend()  #by default, matplotlib tries to determine best location for legend
plt.show()
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, label='x vs y')
axes.plot(x, y**2, label='x vs y^2')
axes.plot(x, y**3, label='x vs y^3')

axes.legend(loc=1)  #1 means upper right. We can also write 'upper right' instead of 1
plt.show()
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, label='x vs y')
axes.plot(x, y**2, label='x vs y^2')
axes.plot(x, y**3, label='x vs y^3')

axes.legend(loc=(1.1, 0.4))  #specifying coordinates manually
# loc = (left, bottom)
# bottom left of legend is at a distance of (1.1 * figure width) from left edge of figure
# bottom left of legend is at a distance of (0.4 * figure height) from bottom edge of figure

plt.show()
```

## Styling

```python
fig, axes = plt.subplots()

# the style of the plot can be specified using format string
# fmt = '[marker][line][color]'  order not important

axes.plot(x, x**2, 'b.-')  #color: blue, marker: ., line: -
axes.plot(x, x ** 3, 'g--') #color: green, line: --

plt.show()
```

## Detailed Styling

```python
fig, axes = plt.subplots()
# color: name of the color or the hex code
# alpha: opacity (1: opaque, 0: transparent)
axes.plot(x, x**2, color="#b54c2f", alpha=0.5)
```

```python
fig, axes = plt.subplots()
# linewidth or lw (both can be used): The width of the line (default=1)
# linestyle or ls (both can be used): The style of the line
axes.plot(x, x**2, lw=5, ls='-.')
plt.show()
```

```python
fig, axes = plt.subplots()
# marker: type of marker used
# markersize: size of marker
axes.plot(x, x**2, marker='o', markersize=20)
plt.show()
```

```python
fig, axes = plt.subplots()
# markerfacecolor: color of the inside color
# markeredgewidth: width of marker border
# markeredgecolor: color of marker border
axes.plot(x, x**2, marker='o', markersize=20, markerfacecolor='blue', markeredgewidth=3, markeredgecolor='r')
plt.show()
```
