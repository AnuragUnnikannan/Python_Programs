## Numpy

```python
import numpy as np
a = np.array([1, 2, 3])  #creating an array using numpy
print(a)
print(type(a))  #printing data type of a
print(a.ndim)  #printing dimension of the array
print(a.shape)
print(len(a)) #printing length of array
np.arange(15).reshape(3, 5)  #filling a matrix with 1st 15 whole numbers
```

```python
import numpy as np
print(np.linspace(1, 8, 4))  #printing numbers in the range spaced equally
print()
print(np.linspace(1, 8, 4, endpoint=False))  #same as abovr except dropping last element
print()
print(np.random.randint(1, 100, 6))  #inserting random numbers in array
print()
print(np.random.randint(1, 100, 6).reshape(3, 2))  #inserting random numbers in a matrix
print()
print(np.random.rand(4))
print()
print(np.eye(5))  #printing identity matrix of 5x5
print()
print(np.zeros)
print()
print(np.random.randn(3, 2))
print()
print(np.empty((3,3)))
print()
print(np.ones((2,3,4), dtype=np.int16))  #printing 2 matrices of 4x3 size with 1's
print()
print(np.array([[1,2],[3,4]],dtype=complex))  #printing complex matrix
print()
print(np.array([(1.5,2,3),(4,5,6)]))
```

```python
A = np.array([[2, 1], [0, 1]])
B = np.array([[2, 0], [3, 4]])
print(A+B)  #sum of 2 matrices
print()
print(A*B)  #element-wise multiplication of matrix
print()
print(A@B)  #matrix multiplication
print()
print(A.dot(B)) #dot product of 2 matrices
print()
print(B.T)  #transpose
print()
print(A.flatten())  #printing matrix in a single dimension array
print()
print(B<3)  #checking whether each element of B is < 3 or not
print()
print(A.sum())  #sum of all elements
print()
print(A.sum(axis=0)) #sum of each column
print()
print(A.sum(axis=1)) #sum of each row
print()
print(A.cumsum(axis=1))  #cumulative sum of rows
print()
print(A.max())  #maximum element of A
print()
print(A.min())  #minimum element of A
print()
```

```python
a = np.arange(4)**3  #inserting cube of all numbers from 0-3
print(a)
print()
print(a[2])
print()
print(a[::-1])  #reversing the array
print()
print(a[0:4:1])
print()
print(a[1,...])
print()
print(a[a>5])  #printing elements of a > 5
print()
x = a[0:4]
print(x)
print()
x[:] = 99  #replacing all elements of x with 99
print(x)
```

## Pandas

```python
import numpy as np
import pandas as pd
arr = np.array([1, 3, 5, 7, 9])
s2 = pd.Series(arr)  #representing an array as series
print(s2)
print(type(s2))
```

```python
import numpy as np
import pandas as pd
s2 = pd.Series([10, 20, 30])  #array as series
print(s2)
print(type(s2))
#array as dataframe
s3 = pd.DataFrame([[1,2],[3,4]], columns=['A','B'], index=['C','D'])
print(s3)
print(type(s3))
```

```python
s4 = pd.DataFrame(np.random.randn(20,7), columns=['A', 'B', 'C', 'D', 'E', 'F', 'G'])
print(s4)
print()
print(s4[(5<s4.index) & (s4.index <10)])  #s4 with values that satisfy both conditions
print()
print(s4.head())  #first 5 rows of s4
print()
print(s4.tail())  #last 5 rows of s4
print()
s4.describe()  #statistical information of data 
print()
print(s4['B'])  #data of 'B' column
print()
print(s4[['B', 'E']])  #data of 'B and 'E' columns
print()
print(s4[0:3])  #data of 1-3 columns
print()
print(s4.iloc[0:3])  #data of 1-3 columns
print()
print(s4.loc[[2,3], ['A', 'B']])  #value of row 2,3 column 'A', 'B'
print()
print(s4[2<s4])  #s4 with values matching conditions
print()
print(s4.mean())  #means of each column
print()
print(s4.mean(1))  #mean of each row
print()
print(s4.drop(5, axis=0))  #delete row 5
print()
print(s4.drop('D', axis=1))  #delete 'D' column
print()
print(s4.drop('D', axis=1, inplace=True))  #delete 'D' column permanently
print()
s4['H'] = np.random.rand(20, 1)  #add a new column of same length.
print(s4)
```
