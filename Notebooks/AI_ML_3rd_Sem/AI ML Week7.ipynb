```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
import numpy as np;
import pandas as pd;
data1 = pd.DataFrame(data = np.c_[iris['data'], iris['target']], columns = iris['feature_names'] + ['target'])
print(data1.head)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
#scatter the sepal_length against sepal_width
ax.scatter(data1['sepal length (cm)'], data1['sepal width (cm)'])
#set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal length')
ax.set_ylabel('sepal width')
```

```python
#get columns to plot
columns = data1.columns.drop(['target'])
#create x data
x_data = range(0, data1.shape[0])
#create figure and axis
fig, ax = plt.subplot()
#plot each column
for column in columns:
    ax.plot(x_data, data1[column], label=column)
#set title and legend
ax.set_title('Iris Dataset')
ax.legend()
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
data['MEDV'] = pd.DataFrame(boston.target)
```

```python
data.head()
```

```python
pd.DataFrame(data.corr().round(2))
```

```python
print(boston.DESCR)
```

```python
x = data['RM']
y = data['MEDV']
```

```python
pd.DataFrame([x, y]).transpose().head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
x_train = pd.DataFrame(x_train)
x_test = pd.DataFrame(x_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

### To measure the performance evaaluation of the model

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
print(sqrt(mean_squared_error(y_pred, y_test)))
```

```python
print("MDEV = {}RM + {}".format(model.coef_[0].round(2), model.intercept_.round(2)))
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x_test, y_test, label='Actual')
plt.plot(x_test, y_pred, color='red', label='Fit')
plt.xlabel('RM')
plt.ylabel('MEDV')
plt.legend()
plt.show()
```
