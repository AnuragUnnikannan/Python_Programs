## Make Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read CSV

```python
df = pd.read_csv('office_sales.csv')
```

```python
df.head()
```

```python
df.info()
```

## Scatterplot

```python
sns.scatterplot(x='salary', y='sales', data=df)
# here salary is along the x-axis and sales is along y-axis
```

```python
# seaborn is built on top of matplotlib, so we can still use all matplotlib functions to modify the figure
plt.figure(figsize=(12, 8), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='division', palette='mako')
# The data points have different color depending on the 'division' column corresponding to that data
# palette = 'mako' sets the color scheme used for this plot.
# if reverse color is required, use 'mako_r'
```

```python
plt.figure(figsize=(12, 8), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, size='training level', alpha=0.7)
```

```python
plt.figure(figsize=(12, 8), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='division', style='division', s=120)
```

## Distribution Plots

Three types of plots:
* Rug
* Histogram
* KDE

```python
plt.figure(figsize=(10, 4), dpi=100)
sns.rugplot(x='salary', data=df, height=0.5)  #heigh: height of each mark (fraction of total height)
```

## Displot (Histogram)

```python
sns.set(style='darkgrid')
# sets grid style for the whole notebook (other: whitegrid, dark, white, ticks)
```

```python
sns.displot(x='salary', data=df, kde=False, rug=False, bins=10)
```

```python
sns.displot(x=df['salary'], kde=True, rug=True, bins=40)
# bins is the number of bars
# rug means the rug plot
```

## KDE (Kernel Density Estimate)

A kernel density estimate plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous  probability density curve in one or more dimensions.

```python
sns.kdeplot(x=df['salary'])
```

```python
sns.kdeplot(x=df['salary'], shade=True, color='#225577')
```

```python
sns.kdeplot(x=df['salary'], clip=(60000, 110000))
```

```python
sns.kdeplot(x='salary', y='sales', data=df)
# with both x and y variables , we get a 2-D KDE plot in the form of contours
```

```python
sns.kdeplot(x=df['salary'], bw_adjust=0.4)
```

## Categorical Plots

### Countplot

It shows the number of occurrences of an item based on certain type of category

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.countplot(x='division', data=df)
```

```python
plt.figure(figsize=(10, 8), dpi=200)
sns.countplot(x='division', data=df, hue='training level', palette='rocket')
```

## Barplot

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.barplot(x='level of education', y='salary', data=df, estimator=np.mean, ci='sd')
```

ci = sd, means it will show the standard deviation

estimator = np.mean, means it will calcultate the mean salary for each category

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.barplot(x='level of education', y='salary', data=df, hue='training level', estimator=np.mean, ci='sd')
```

## Distribution Within Categories

### Box Plot

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.boxplot(x='math score', data=df)
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.boxplot(x='parental level of education', y='math score', data=df)
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.boxplot(y='parental level of education', x='math score', hue='gender', data=df)
```

## Violin Plot

```python
plt.figure(figsize=(6, 12), dpi=100)
sns.violinplot(x='math score', y='parental level of education', data=df, hue='gender', 
               split=True, inner='quartile', bw=0.2)
# the violin plots are symmetrical so using split=True along with hue, we can have two halves of the plot can show different values.
# bw is the bandwidth (equivalent to bw_adjust of kdeplot)
```

## Swarmplot

Non overlapping scatter plot

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.swarmplot(x='math score', data=df)
```

```python
plt.figure(figsize=(10, 10), dpi=100)
sns.swarmplot(x='math score', y='race/ethnicity', data=df)
```

```python
plt.figure(figsize=(10, 10), dpi=100)
sns.swarmplot(x='math score', y='race/ethnicity', hue='gender', data=df)
```

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.swarmplot(y='math score', x='race/ethnicity', data=df, hue='gender', dodge=True, size=4)
# We set dodge=True to separate the hue
```

## Boxenplot

```python
plt.figure(figsize=(10, 6), dpi=100)
sns.boxenplot(y='math score', x='race/ethnicity', data=df, hue='gender')
```

## Matrix Plots

A matrix plot is a color-coded diagram that has rows data columns data, and values

```python
df = pd.read_csv('country_table.csv')
df = df.set_index('Countries')
df.head()
```

```python
plt.figure(figsize=(5, 5), dpi=100)
sns.heatmap(df)
```

```python
sns.heatmap(df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='mako')
```

## Clustermap
