### Question 1:  Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.

```python
import matplotlib.pyplot as plt
from pylab import randn
x = randn(100)
y = randn(100)
plt.figure(figsize=(10, 7))
plt.scatter(x, y, color='blue')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### Question 2: Write a Python program to draw a scatter plot using random distributions to generate balls of different sizes

```python
import math 
import random 
import matplotlib.pyplot as plt 
no_of_balls = 25 
x = [random.triangular() for i in range(no_of_balls)] 
y = [random.gauss(0.5, 0.25) for i in range(no_of_balls)] 
colors = [random.randint(1, 4) for i in range(no_of_balls)] 
areas = [math.pi * random.randint(5, 15)**2 for i in range(no_of_balls)] 
plt.figure(figsize=(10,7)) 
plt.scatter(x, y, s=areas, c=colors, alpha=0.85) 
plt.axis([0.0, 1.0, 0.0, 1.0]) 
plt.title('Generate balls of different sizes') 
plt.xlabel("X_Data") 
plt.ylabel("Y_Data") 
plt.show()
```

### Question 3: Write a program to plot a bar chart (histogram) in python to display the result of any column of Iris Dataset.

```python
from sklearn import datasets
iris = datasets.load_iris()
```

```python
import numpy as np
import pandas as pd
data1 = pd.DataFrame(data = np.c_[iris['data'], iris['target']], columns = iris['feature_names'] + ['target'])
```

```python
list(data1.columns)
```

```python
data1.describe()
```

```python
plt.figure(figsize = (10, 7))
x = data1["sepal width (cm)"]
plt.hist(x, bins=20, color="orange")
plt.title("Histogram")
plt.xlabel("Sepall Width(cm)")
plt.ylabel("Count")
```

###  Question 4: Import a dataset from the internet and plot line graph for 3 of the columns denoted with different colors.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
rain = pd.read_csv(r'rainfall3.csv')
x1 = [2011, 2012, 2013, 2014, 2015]
y1 = rain['UTTARAKHAND']
y2 = rain['HARYANA']
y3 = rain['PUNJAB']
plt.figure(figsize = (12, 7))
plt.plot(x1, y1, marker = 'o')
plt.plot(x1, y2, marker = 'o')
plt.plot(x1, y3, marker = 'o')
plt.xlabel('YEAR')
plt.ylabel('RAINFALL (cm)')
plt.show()
```

### Question 5: Write a Python program to draw a box plot graph taking a random distribution in X and Y and plotted against each other.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize = (10, 7))
plt.boxplot(data)
plt.show()
```
