```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(diabetes.data, columns = diabetes.feature_names)
```

```python
data['Target'] = pd.DataFrame(diabetes.target)
```

```python
data.head()
```

```python
pd.DataFrame(data.corr().round(2))
```

```python
print(diabetes.DESCR)
```

```python
x = data[['bmi', 's5', 'bp']]
y = data['Target']
```

```python
pd.DataFrame([x, y]).transpose().head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
x_train = pd.DataFrame(x_train)
y_train = pd.DataFrame(y_train)
x_test = pd.DataFrame(x_test)
y_test = pd.DataFrame(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

### To measure the performance evaluation of the model

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
print(sqrt(mean_squared_error(y_pred, y_test)))
```

```python
print("Target = {}bmi + {}".format(model.coef_[0].round(2), model.intercept_.round(2)))
```
